```python
  import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("Project_emp_attrition_dataset.csv")

# See the shape — how many rows and columns
print("Shape:", df.shape)

# See first 5 rows
print(df.head())

# See all column names and their data types
print(df.info())

# Basic statistics for numerical columns
print(df.describe())
```

```python
# Drop the empty column and columns with only one unique value (useless for analysis)
df = df.drop(columns=['Unnamed: 35', 'EmployeeCount', 'StandardHours', 'Over18'])

print("Columns after dropping:", df.shape)
print("Remaining columns:", list(df.columns))
```

```python
# Check duplicates
print("Duplicate rows:", df.duplicated().sum())

# Remove if any
df.drop_duplicates(inplace=True)
print("Shape after removing duplicates:", df.shape)
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Standardize text columns — remove spaces, fix capitalization
df['Attrition'] = df['Attrition'].str.strip().str.title()
df['Gender'] = df['Gender'].str.strip().str.title()
df['Department'] = df['Department'].str.strip().str.title()
df['OverTime'] = df['OverTime'].str.strip().str.title()
df['MaritalStatus'] = df['MaritalStatus'].str.strip().str.title()
df['BusinessTravel'] = df['BusinessTravel'].str.strip()

# Verify unique values
print("Attrition:", df['Attrition'].unique())
print("Gender:", df['Gender'].unique())
print("Department:", df['Department'].unique())
print("OverTime:", df['OverTime'].unique())
print("MaritalStatus:", df['MaritalStatus'].unique())
```

```python
# Age Group column
df['AgeGroup'] = pd.cut(df['Age'],
                         bins=[18, 25, 35, 45, 60],
                         labels=['18-25', '26-35', '36-45', '46-60'])

# Salary Category column
df['SalaryCategory'] = pd.cut(df['MonthlyIncome'],
                               bins=[0, 3000, 6000, 10000, 100000],
                               labels=['Low', 'Medium', 'High', 'Very High'])

print(df[['Age', 'AgeGroup', 'MonthlyIncome', 'SalaryCategory']].head(10))
```

```python
# Print cleaning summary
print("========== CLEANING SUMMARY ==========")
print(f"Total Rows        : {df.shape[0]}")
print(f"Total Columns     : {df.shape[1]}")
print(f"Missing Values    : {df.isnull().sum().sum()}")
print(f"Duplicate Rows    : {df.duplicated().sum()}")
print(f"New Columns Added : AgeGroup, SalaryCategory")
print("=======================================")

# Save cleaned file
df.to_csv("employee_attrition_analysis.csv", index=False)
print("✅ Cleaned file saved!")
```

```python
# Find exactly which columns have missing values
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# Fix AgeGroup missing values
df['AgeGroup'] = df['AgeGroup'].astype(str)
df['AgeGroup'].fillna('Unknown', inplace=True)

# Fix SalaryCategory missing values
df['SalaryCategory'] = df['SalaryCategory'].astype(str)
df['SalaryCategory'].fillna('Unknown', inplace=True)

# For any other numerical column missing values
for col in df.select_dtypes(include='number').columns:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

# For any other text column missing values
for col in df.select_dtypes(include='object').columns:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)

print("Missing values after fix:", df.isnull().sum().sum())
```
